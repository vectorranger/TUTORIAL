## Step 1: Create Your Workspace

Every notebook in this tutorial reads and writes data relative to a single folder called `EHR_TUTORIAL_WORKSPACE`. We do this so that paths stay consistent across all notebooks — you set it once and everything works.

Here is the target directory structure:

```
EHR_TUTORIAL_WORKSPACE/
├── raw_data/
│   ├── structured_data/   ← MIMIC-IV hospital & ICU tables
│   └── note_data/         ← MIMIC-IV clinical notes
├── processed_data/        ← outputs from cleaning & analysis
└── scripts/               ← tutorial notebooks & mapping files
```

You can place `EHR_TUTORIAL_WORKSPACE` anywhere on your system. The code below creates it in your current working directory.

````python
import os

base_dir = "EHR_TUTORIAL_WORKSPACE"

for folder in ["raw_data/structured_data", "raw_data/note_data", "processed_data", "scripts"]:
    os.makedirs(os.path.join(base_dir, folder), exist_ok=True)

print(f"Workspace created at: {os.path.abspath(base_dir)}")
````

Let's confirm the directory structure was created correctly.

````python
for dirpath, dirnames, filenames in os.walk(base_dir):
    level = dirpath.replace(base_dir, "").count(os.sep)
    indent = "  " * level
    print(f"{indent}{os.path.basename(dirpath)}/")
````

You should see the four subdirectories listed above. This structure stays the same throughout the tutorial.

## Step 4: Download MIMIC-IV Data

We need two datasets from PhysioNet, both requiring your credentialed account:

| Dataset | Contents | Destination folder |
|---------|----------|-------------------|
| **MIMIC-IV 3.1** | Structured hospital & ICU tables | `raw_data/structured_data/` |
| **MIMIC-IV-Note 2.2** | Clinical notes (discharge summaries) | `raw_data/note_data/` |

Choose one of the two download methods below.

### Option A: Download with wget (Mac/Linux)

The `wget` command downloads files recursively from PhysioNet. Replace `YOUR_USERNAME` and `YOUR_PASSWORD` with your PhysioNet credentials.

```bash
# Download structured data (hosp and icu modules)
wget -r -N -c -np \
  --user YOUR_USERNAME --password YOUR_PASSWORD \
  https://physionet.org/files/mimiciv/3.1/ \
  -P EHR_TUTORIAL_WORKSPACE/raw_data/structured_data

# Download clinical notes
wget -r -N -c -np \
  --user YOUR_USERNAME --password YOUR_PASSWORD \
  https://physionet.org/files/mimic-iv-note/2.2/ \
  -P EHR_TUTORIAL_WORKSPACE/raw_data/note_data
```

What the flags mean: `-r` downloads recursively, `-N` skips files already downloaded, `-c` resumes interrupted downloads, `-np` stays within the target directory.

> **Windows:** `wget` is not installed by default. Use Option B below, or install wget via `winget install GnuWin32.Wget`.

### Option B: Manual Download from PhysioNet (All OS)

1. Go to the [MIMIC-IV 3.1 files page](https://physionet.org/content/mimiciv/3.1/) and log in
2. Download all `.csv.gz` files from the `hosp/` and `icu/` folders
3. Place them in:
   - `EHR_TUTORIAL_WORKSPACE/raw_data/structured_data/physionet.org/files/mimiciv/3.1/hosp/`
   - `EHR_TUTORIAL_WORKSPACE/raw_data/structured_data/physionet.org/files/mimiciv/3.1/icu/`
4. Go to the [MIMIC-IV-Note 2.2 files page](https://physionet.org/content/mimic-iv-note/2.2/) and log in

The nested `physionet.org/files/...` path structure is created automatically by wget. If you download manually, you need to create these subdirectories so the tutorial paths work correctly.

Let's create the nested data directories now so they are ready for your downloaded files.

````python
data_dirs = [
    os.path.join("raw_data", "structured_data", "physionet.org", "files", "mimiciv", "3.1", "hosp"),
    os.path.join("raw_data", "structured_data", "physionet.org", "files", "mimiciv", "3.1", "icu"),
    os.path.join("raw_data", "note_data", "physionet.org", "files", "mimic-iv-note", "2.2", "note"),
]

for d in data_dirs:
    os.makedirs(os.path.join(base_dir, d), exist_ok=True)
    print(f"  Created: {d}")

print("\nData directories are ready. Place your downloaded files in the paths above.")
````

## Step 5: Unzip the Data

MIMIC-IV files are distributed in **gzip** format (`.csv.gz`) to save download time and disk space. We need to decompress them into regular `.csv` files before our notebooks can read them.

We use Python's built-in `gzip` and `shutil` modules, which work on every operating system without requiring any extra software.

Let's first see how decompression works on a single file.

````python
import gzip
import shutil

# Pick one file to demonstrate the decompression steps
example_gz = os.path.join(base_dir, "raw_data", "structured_data",
                          "physionet.org", "files", "mimiciv", "3.1",
                          "hosp", "patients.csv.gz")

output_csv = example_gz.replace(".gz", "")

with gzip.open(example_gz, "rb") as f_in:
    with open(output_csv, "wb") as f_out:
        shutil.copyfileobj(f_in, f_out)

print(f"Decompressed: {os.path.basename(output_csv)}")
````

We opened the compressed file with `gzip.open`, then copied its contents to a new file without the `.gz` extension using `shutil.copyfileobj`. Now let's apply this to all compressed files in our data directories.

````python
import glob

gz_files = glob.glob(os.path.join(base_dir, "raw_data", "**", "*.gz"), recursive=True)
print(f"Found {len(gz_files)} compressed files to decompress.\n")

for gz_path in gz_files:
    output_path = gz_path.replace(".gz", "")
    if os.path.exists(output_path):
        print(f"  Skipping (already decompressed): {os.path.basename(output_path)}")
        continue
    with gzip.open(gz_path, "rb") as f_in:
        with open(output_path, "wb") as f_out:
            shutil.copyfileobj(f_in, f_out)
    print(f"  Decompressed: {os.path.basename(output_path)}")

print("\nDone. All files decompressed.")
````

Notice that the code skips files that are already decompressed. This means you can safely re-run this cell without duplicating work.

## Step 6: Verify Your Setup

Before moving on, let's confirm that the key data files are in place.

````python
expected_files = [
    os.path.join("raw_data", "structured_data", "physionet.org", "files", "mimiciv", "3.1", "hosp", "patients.csv"),
    os.path.join("raw_data", "structured_data", "physionet.org", "files", "mimiciv", "3.1", "hosp", "diagnoses_icd.csv"),
    os.path.join("raw_data", "structured_data", "physionet.org", "files", "mimiciv", "3.1", "hosp", "labevents.csv"),
    os.path.join("raw_data", "note_data", "physionet.org", "files", "mimic-iv-note", "2.2", "note", "discharge.csv"),
]

all_found = True
for filepath in expected_files:
    full_path = os.path.join(base_dir, filepath)
    exists = os.path.exists(full_path)
    status = "Found" if exists else "MISSING"
    print(f"  [{status}] {filepath}")
    if not exists:
        all_found = False

if all_found:
    print("\nAll key files are in place. You are ready to start the tutorial!")
else:
    print("\nSome files are missing. Check the download and unzip steps above.")
````

If all files show `[Found]`, your setup is complete. If any show `[MISSING]`, revisit Steps 4 and 5 to ensure the data was downloaded and decompressed into the correct directories.

## Advanced: Running on an HPC Cluster with SLURM

> **This section is optional.** If you are running the tutorial on a laptop or desktop, skip ahead to **Next Step** below. This section is for users who want to run the tutorial notebooks on a high-performance computing (HPC) cluster managed by SLURM.

### When Do You Need HPC?

Most of this tutorial runs comfortably on a laptop with 8–16 GB of RAM. However, two operations benefit from HPC resources:

- **Processing `labevents.csv`** (100M+ rows) in the Getting Started and Data Cleaning notebooks
- **NLP processing of clinical notes** in Notebook 3

If your machine runs out of memory or a notebook takes too long, submit it as a batch job on your cluster.

### Setting Up Your Environment on HPC

Most HPC clusters use environment modules to manage software. Load Python and create a virtual environment inside your workspace:

```bash
# Load Python (module name varies by cluster — check your docs)
module load python/3.10

# Create and activate a virtual environment
cd EHR_TUTORIAL_WORKSPACE
python -m venv ehr_env
source ehr_env/bin/activate

# Install dependencies
pip install -r scripts/EHR-Processing-Tutorial/requirements.txt
```

Some clusters use `module load anaconda` instead. Check your cluster's documentation for the exact command.

### SLURM Batch Job Template

Save the following as `run_notebook.sbatch` in your `scripts/` directory. It converts a Jupyter notebook to a Python script and executes it.

```bash
#!/bin/bash
#SBATCH --job-name=ehr_tutorial
#SBATCH --partition=short
#SBATCH --time=04:00:00
#SBATCH --mem=32G
#SBATCH --cpus-per-task=4
#SBATCH --output=ehr_tutorial_%j.log
#SBATCH --error=ehr_tutorial_%j.err

# Load Python and activate environment
module load python/3.10
source EHR_TUTORIAL_WORKSPACE/ehr_env/bin/activate

# Convert notebook to Python script and run it
NOTEBOOK=$1
jupyter nbconvert --to script "$NOTEBOOK" --output temp_script
python temp_script.py

echo "Job complete: $NOTEBOOK"
```

Key SLURM directives:
- `--mem=32G`: Request enough RAM for large files like `labevents.csv`
- `--time=04:00:00`: Wall clock limit (4 hours is generous for any single notebook)
- `--partition=short`: Change this to match your cluster's available partitions

### Submitting a Job

```bash
sbatch run_notebook.sbatch EHR_Preprocessing_Getting_Started.ipynb
```

Monitor your job with `squeue -u $USER`. Output logs appear in `ehr_tutorial_<jobid>.log`.

**Tip:** For the labs-heavy notebooks (Getting Started, Data Cleaning), request at least 32 GB of memory. For NLP (Notebook 3), 16 GB is usually sufficient.
